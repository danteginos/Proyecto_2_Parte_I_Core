```python
# # Crear la figura y los subplots.
fig, axs = plt.subplots(2,4, figsize=(14, 6))
# Se deja el array en una dimension para no usar indices con dos dimensiones
axs = axs.flatten()
# Columnas num√©ricas a graficar
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns.to_list()
# Boxplot para las columnas num√©ricas
prop_caja = dict(facecolor='skyblue')
for i, col in enumerate(numeric_cols[:15]):
  axs[i].boxplot(df[col], patch_artist=True, boxprops=prop_caja)
  axs[i].set_title(f"Boxplot {col}", fontsize=10, fontweight='bold')
  axs[i].set_xticklabels([col], fontweight='bold')
  axs[i].set_ylabel("Valores", fontweight='bold')
  axs[i].grid(True, linewidth=0.5)

```

#### Se encuentran datos atipicos en columna avg_watch_time_per_day y en wath_hours. Los cuales estan alejados del resto de los datos. Se puede suponer que son casos reales quwe se dan y dejarlos en el procesamiento o eliminarlos.

```python
# Se crea el histograma para visualizar los datos atipicos.
plt.hist(df["watch_hours"], bins=10, edgecolor='black')
plt.title('Histograma de Horas Vistas')
plt.xlabel('watch_hours')
plt.ylabel('Frecuencia')
plt.show()
```

#### La horas vistas se encuentran distribuidas entre 0 y 60 horas, con una gran frecuencia en el rango de 0 a 20 horas

```python
# Se crea el histograma para visualizar los datos atipicos.
plt.hist(df["avg_watch_time_per_day"], bins=10, edgecolor='black')
plt.title('Histograma Tiempo promedio de visualizaci√≥n por dia ')
plt.xlabel('avg_watch_time_per_day')
plt.ylabel('Frecuencia')
plt.show()
```

#### El tiempo promedio de visualizaciones por dia se concentra en el rango de 0 a 20 y con una frecuencia muy alta, cerca de 500.
